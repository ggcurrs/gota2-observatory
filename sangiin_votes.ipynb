```
dict_excel_url = {
  "北海道": "https://www.soumu.go.jp/main_content/000637555.xlsx", 
  "東京都": "https://www.soumu.go.jp/main_content/000637591.xlsx",  
  "沖縄県": "https://www.soumu.go.jp/main_content/000637693.xlsx"
}
```

## Funcions


Function 1: Get an Excel file from a given URL and return a pandas.DataFrame

```
import requests
import pandas as pd
import time


def excel2df(excel_url):
  """Get an Excel file and return a pandas.DataFrame"""
  try:
    response = requests.get(excel_url)  # Download
    excel_file = response.content
    df_excel = pd.read_excel(excel_file)  # Conv to a DF

    return df_excel

  except Exception as exc:
    print(f'Error: {exc}')
```

Function 2: Clean up the DataFrame

```
def df_cleaner(my_df):
  """Clean up the DataFrame"""
  # Extract a row containing candidate names, conv to a list.
  new_col_names = my_df.loc[2, :].to_list()

  # Make a new DataFrame
  df_new = my_df.copy()  # 'Copy' in order to keep the original intact.
  df_new.columns = new_col_names

  # Remove unnecessary rows and columns
  df_new.drop([0, 1, 2, 3], inplace=True)
  df_new.dropna(axis='columns', inplace=True)

  # Reset the index
  df_new.reset_index(drop=True, inplace=True)


  return df_new
```

Function 3: Makes cucumber pickles as 'basename.pickle'

```
import pickle

def vinegar(cucumber, base_name):
  """Makes cucumber pickles as 'basename.pickle'"""
  with open(base_name+'.pickle', 'wb') as f:
    pickle.dump(cucumber, f)
  
  print(
   'FYI: The pickle module is not secure. Only unpickle data you trust.'
  )

  return 0
```

## Main
- Makes a dictionary of DFs with district names as keys.

```
dict_df = {}  # Key => District, Value => DataFrame

for each_key in dict_excel_url:
  each_df = excel2df(dict_excel_url[each_key])  # Using Function 1
  each_df = df_cleaner(each_df)  # Using Function 2

  # Append to the dictionary.
  dict_df[each_key] = each_df  # {<district name>: <DataFrame>}

  # Progress report
  print(
      '\nKey: ', each_key,
      '\nURL: ', dict_excel_url[each_key],
      '\nDF : ', each_df.shape, '(rows, columns)',
      '\nPose for 5 sec.'
  )

  time.sleep(5)

print('\n\nDONE!')


# Save the resultant DF dictionary.
vinegar(dict_df, 'sangiin')  # Using Function 3

print('\nSaved: sangiin.pickle')
```
